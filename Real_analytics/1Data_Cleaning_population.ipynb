##### We found there was country name called "not classified" where there was not data in it from year 1960 to 2018 so we removed it whole

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# List of known values (1960–2011)
population_values = [
    1397491, 1432640, 1469645, 1508273, 1548187, 1589179, 1631147, 1674204, 1718525,
    1764343, 1811878, 1861199, 1912302, 1965160, 2019717, 2075965, 2133723, 2193068,
    2254450, 2318495, 2385540, 2454766, 2525521, 2598410, 2674289, 2753151, 2837111,
    2924349, 3006361, 3071771, 3113311, 3127297, 3118582, 3099047, 3085443, 3090159,
    3116379, 3161350, 3224223, 3302263, 3392801, 3497124, 3614639, 3738265, 3858623,
    3969007, 4066648, 4153332, 4232636, 4310334, 4390840, 4474690
]

years_known = np.arange(1960, 2012)  # 1960 to 2011
years_missing = np.arange(2012, 2018)  # 2012 to 2017

# Fit a 2nd-degree polynomial (quadratic trend line)
coeffs = np.polyfit(years_known, population_values, deg=2)
poly = np.poly1d(coeffs)

# Predict missing values
predicted_values = poly(years_missing).astype(int)
print("Predicted values from 2012 to 2017:", predicted_values)

# (Optional) Plot
plt.plot(years_known, population_values, label="Known", marker='o')
plt.plot(years_missing, predicted_values, label="Predicted", marker='x')
plt.legend()
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Population Prediction for Ateriya")
plt.grid(True)
plt.show()
```

```python
years_to_fill = [str(year) for year in range(2012, 2018)]

# Locate the row for Eritrea (country code "ERI")
eritrea_index = population_df[population_df['Country Code'] == 'ERI'].index

# Assign the predicted values to the appropriate columns
for i, year in enumerate(years_to_fill):
    population_df.at[eritrea_index[0], year] = predicted_values[i]
```

##### so look at the above simple regression this works for all the data hence we are gonna replace it simply it is more better approach to use prediction 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Define year columns
year_columns = [str(year) for year in range(1960, 2018)]

# Loop through each row (country) and predict missing values
for idx, row in population_df.iterrows():
    # Get available years and corresponding values
    available_years = []
    available_values = []
    missing_years = []

    for year in year_columns:
        value = row[year]
        if pd.isnull(value):
            missing_years.append(int(year))
        else:
            available_years.append(int(year))
            available_values.append(value)

    # Proceed only if there are missing years and enough data to train
    if missing_years and len(available_values) >= 2:
        # Fit a linear regression model
        model = LinearRegression()
        model.fit(np.array(available_years).reshape(-1, 1), available_values)

        # Predict for missing years
        predicted_values = model.predict(np.array(missing_years).reshape(-1, 1))

        # Make all predicted values positive by taking their absolute value and rounding them to nearest integer
        predicted_values = np.abs(predicted_values)
        predicted_values = np.round(predicted_values).astype(int)

        # Update the DataFrame with predicted values
        for year, pred in zip(missing_years, predicted_values):
            population_df.at[idx, str(year)] = pred

        # Plotting the available data and predicted data for countries with missing data
        plt.figure(figsize=(10, 6))
        plt.plot(available_years, available_values, 'o-', label='Available Data', color='blue')
        plt.plot(missing_years, predicted_values, 'x--', label='Predicted Data', color='red')

        # Use the country name for the plot title
        plt.title(f"Population Prediction for Country {row['Country Name']}")
        plt.xlabel('Year')
        plt.ylabel('Population')
        plt.legend()
        plt.grid(True)
        plt.show()
```
